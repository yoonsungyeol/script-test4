```python
def cmp_by_each_element(a):
    place = a.find(' ')
    return a[place+1]
f= open('s.txt','r')
line= f.readlines()
line.sort(key=cmp_by_each_element)
f.close()

f1 = open('s2.txt','w')
f1.writelines(line)
f1.close()
f1 = open('s2.txt','r')
print(f1.read())
f1.close()
```

# 3번문제

    1) range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.
      range(stop)
      range(start, stop[, step])
    2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다.
    3) frange 사용 예
      3-1) frange(0.5)
        [0.0, 0.1, 0.2, 0.3, 0.4]
      3-2) frange(1.0, 2.0)
        [1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]
      3-3) frange(2.2, 4.0, 0.5)
        [2.2, 2.7, 3.2, 3.7]

```python
def frange(final,start=0.0,step=0.1):
    l = []
    while(True):
        if start >=final:
            break
        l.append(start)
        start += step
    print(l)
frange(0.5)
frange(2.0,1.0)
frange(4.0, 2.2, 0.5)
```

    <설명>
    - frange라는 함수를 정의해 주는데 인자로 final값만 받아도 가능하도록 하기 위해 인자에 start를 0.0으로 step을 0.1로 하여 설정해 줍니다.(여기서 final이 첫자리에 오지 않으면 오류가 나므로 첫자리로 final이 오도록 설정합니다.)
    - 새로운 리스트l을 선언하고 start값부터 시작해서 step만큼 증가되는 값들을 모두 리스트l에 추가해 줍니다.
    - 증가된 start값이 final을 넘게 되면 저장을 멈추도록 하고 리스트를 출력하여 함수정의를 마무리합니다.
    - 마지막으로 결과가 어떻게 나오는지 함수에 수를 대입하여 결과를 확인합니다.

## 5번문제

   - 여러 단어로 이루어진 문자열을 입력받아 각 단어의 첫글자로 이루어진 단어를 대문자로 출력하는 myinitial() 함수를 다음 조건을 참고하여 작성하시오.
   
     - 1) 다음에 제시되는 함수들을 모두 이용해야 한다.
           split
           map
           join
       2) myinitial() 함수 사용 예
           myinitial("as soon as possible")
              ASAP

```python
def myinitial(s):
    s2 =s.title()
    s3 = s2.split()
    s4 = list(map(lambda x: x[0],s3))
    print(''.join(s4))
myinitial("as soon as possible")
```

    <설명>
    - myinitial함수를 정의하는데 입력된 문자열의 첫단어들을 대문자로 바꾸기 위해 title()을 이용하고 바꾼 문자열을 s2에 저장합니다.
    - s2를 split()으로 리스트로 만들어 s3에 저장합니다.
    - lambda를 이용하여 각 리스트 원소의 첫 단어를 빼와서 map으로 합쳐주고 list()로 묶어서 s4에 저장합니다.
    - s4를 join()으로 합쳐서 출력합니다.

## 6번문제

   - 음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.

```python
num = int(input())
def myfact(n):
    if n == 0:
        return 1
    return n*myfact(n-1)
print(myfact(num))
```

    <설명>
    - 숫자를 input()으로 입력받습니다.
